Project: **Mall Customer Segmentation using K means clustering**

Algorithm Used: **K-Means Clustering**

Steps:




*   Importing Necessary Libraries.
*   Selecting the variables to be clustered.


*   Plotting Elbow Graph using WCSS to find the optimal number of clusters.
*   Clustering using K-Means










```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
df=pd.read_csv("/content/Mall_Customers.csv")
df
```

```python
X=df.iloc[:,3:]
X
```

```python
from sklearn.cluster import KMeans
```

```python
#plotting elbow graph
wcss=[]
for i in range(1,9):
  km=KMeans(n_clusters=i,init='k-means++',random_state=4)
  km.fit(X)
  wcss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,9),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km1=KMeans(n_clusters=5,init='k-means++',random_state=4)
y_kmeans=km1.fit_predict(X)
y_kmeans
```

```python
X=np.array(X)
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1],s = 100,c = 'red', label = 'Cluster1')
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1],s = 100,c = 'blue', label = 'Cluster2')
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1],s = 100,c = 'green', label = 'Cluster3')
plt.scatter(X[y_kmeans==3,0], X[y_kmeans==3,1],s = 100,c = 'pink', label = 'Cluster4')
plt.scatter(X[y_kmeans==4,0], X[y_kmeans==4,1],s = 100,c = 'lime', label = 'Cluster5')
plt.scatter(km1.cluster_centers_[:,0], km1.cluster_centers_[:,1], s = 100, c = 'black', label = 'Centroid')
plt.title("Clusters")
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
X= pd.DataFrame(X,columns=('Annual Income (k$)','Spending Score (1-100)'))
y=np.array(y_kmeans)
X['target']=y
X
```

Based on the clusters formed customers can be classified as:

*   Cluster 1- Average income - Average spending score
*   Cluster 2- Low income - Low spending score


*   Cluster 3- High income - Low spending score
*   Cluster 4- Low income - High spending score

*   Cluster 5- High income - High spending score
